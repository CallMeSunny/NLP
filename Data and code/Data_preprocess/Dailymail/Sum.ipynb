```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#import the pakeges
```

```python
y1 = np.load('Belt_pos.npy')/3 + np.load('Community_pos.npy')/3 + np.load('Dream_pos.npy')/3
y2 = np.load('Belt_neg.npy')/3 + np.load('Community_neg.npy')/3 + np.load('Dream_neg.npy')/3
```

```python
group_labels = np.load('labels.npy')
```

```python
x1 = np.arange(0,71)
x2 = x1
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig,ax = plt.subplots()

plt.figure(figsize=(38, 22))
plt.plot(x1,y1,lw = 3)
plt.plot(x2,y2,lw = 3)
plt.legend(['Positive Line','Negtive Line'],fontsize = 'xx-large')
my_x_ticks = np.arange(0, 71)
plt.xticks(my_x_ticks, group_labels, rotation=0)
plt.grid()
plt.title('The Curves in Dailymail.co.uk',fontsize = 'xx-large')
plt.xlabel('Time since 2013.06',fontsize = 'xx-large')
plt.ylabel('The factors (0~1)',fontsize = 'xx-large')
plt.savefig('Dailymail.jpg')
plt.show()
```
